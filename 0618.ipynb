```python
BTC_input["buy_day"]  = 0
BTC_input["cum_day"]  = 0
BTC_input["sell_day"] = 0
BTC_input["RSV"] = 0
BTC_input["K"] = 0
BTC_input["D"] = 0
```

```python
for i in range(9, BTC_input.shape[0]):
    close=BTC_input["Close"].iloc[i]
    high=BTC_input["High"].iloc[i]
    low=BTC_input["Low"].iloc[i]
    for j in range(1,9):
        if(BTC_input["High"].iloc[i-j]>high):
            high=BTC_input["High"].iloc[i-j]
        if(BTC_input["Low"].iloc[i-j]<low):
            low=BTC_input["Low"].iloc[i-j]
    BTC_input["RSV"].iloc[i]=(close-low)/(high-low)*100
```

```python
BTC_input["K"].iloc[9]=50
BTC_input["D"].iloc[9]=50
```

```python
#K
for i in range(10, BTC_input.shape[0]):
    k=BTC_input["K"].iloc[i-1]*2/3+BTC_input["RSV"].iloc[i]*1/3
    BTC_input["K"].iloc[i]=k
#D
for i in range(10, BTC_input.shape[0]):
    d=BTC_input["D"].iloc[i-1]*2/3+BTC_input["K"].iloc[i]*1/3
    BTC_input["D"].iloc[i]=d
```

```python
#黃金與死亡交叉 買賣訊號
for i in range(10, BTC_input.shape[0]):
    if(BTC_input["K"].iloc[i]>BTC_input["D"].iloc[i]) and (BTC_input["cum_day"].iloc[i-1]==0):
        BTC_input["buy_day"].iloc[i]=1
        BTC_input["cum_day"].iloc[i]=1
        BTC_input["sell_day"].iloc[i]=0
    elif(BTC_input["K"].iloc[i]<BTC_input["D"].iloc[i])and (BTC_input["cum_day"].iloc[i-1]==1):
        BTC_input["buy_day"].iloc[i]=0
        BTC_input["cum_day"].iloc[i]=0
        BTC_input["sell_day"].iloc[i]=1
    elif(BTC_input["cum_day"].iloc[i-1]>0):
        BTC_input["buy_day"].iloc[i]=0
        BTC_input["cum_day"].iloc[i]=1
        BTC_input["sell_day"].iloc[i]=0
```

```python
initial = 10000
for i in range(0, BTC_input.shape[0]):
#     買的訊號
    if(BTC_input["K"].iloc[i]>BTC_input["D"].iloc[i]) and (BTC_input["cum_day"].iloc[i-1]==0):
#         持有數量 = 手中的資本/ 該標的收盤價
        BTC_input.iloc[i, -3] = initial/ BTC_input["Adj Close"].iloc[i]
#     市值 = 持有數量* 該標的收盤價
        BTC_input.iloc[i, -2] = BTC_input["number"].iloc[i]*BTC_input["Adj Close"].iloc[i]
#     期間報酬 = (市值-投入成本)/投入成本
        BTC_input.iloc[i, -1] = round((BTC_input["value"].iloc[i]-initial)/initial,4)
#         賣的訊號
    elif(BTC_input["K"].iloc[i]<BTC_input["D"].iloc[i])and (BTC_input["cum_day"].iloc[i-1]==1):
#         持有數量為 0 
        BTC_input.iloc[i, -3] = 0
#     市值 = 持有數量* 該標的收盤價
        BTC_input.iloc[i, -2] = BTC_input["number"].iloc[i-1]*BTC_input["Adj Close"].iloc[i]
#     期間報酬 = (市值-投入成本)/投入成本
        BTC_input.iloc[i, -1] = round((BTC_input["value"].iloc[i]-initial)/initial,4)
#         持有的訊號
    elif(BTC_input["cum_day"].iloc[i-1]>0):
#         持有數量與上一期相同
        BTC_input.iloc[i, -3] = BTC_input.iloc[i-1, -3]
#     市值 = 持有數量* 該標的收盤價
        BTC_input.iloc[i, -2] = BTC_input["number"].iloc[i]*BTC_input["Adj Close"].iloc[i]
#     期間報酬 = (市值-投入成本)/投入成本
        BTC_input.iloc[i, -1] = round((BTC_input["value"].iloc[i]-initial)/initial,4)
```
